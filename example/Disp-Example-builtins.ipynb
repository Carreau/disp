# Disp Example

This is an example on how you can use disp to make the environement you use richer.

Let's have a quick look at a nor mal workflow, and how objects are show in the notebook: 

```python
from IPython.terminal.interactiveshell import InteractiveShell
import disp
InteractiveShell
```

```python
print
```

```python
disp
```

This is a bit boring, and not always informative. Let's enable disp for builtin types. We don't do it by default, as it can be heavily computation intensive and does nto work outside of IPython. It will also make your notebook files way larger. 

```python
import disp
disp.activate_builtins()
```

Let's have a look again at previous objects:

```python
InteractiveShell
```

You should see now (at least if you are in an interactive environement) a small triangle in front of `IPython.core.interactiveshell.InteractiveShell`, if you click on it, the output will expand to show you the documentation as well as the inheritence (Actually the MRO... but details for now).

```python
print
```

Print wil have the same treatment, with an overview of the documetation. You can still use `?`/`??` to get more information. 

```python
disp
```

If modules have docstrings, they will be shown as well. 

##  Requests

Above are generic objects, `disp` also have special casing for some objects, example `requests` replies. 

```python
import requests
import requests_cache
requests_cache.install_cache('tmpcache.db')

res = requests.get('https://api.github.com/aoisehgd')
```

`get` is still a generic function, so will show its doc:

```python
requests.get # this is a function, alerady covered by activate_builtins()
```

By default again the repr is not enabled:

```python
res
```

```python
type(res) # this is a type, already cover by activate_builtins()
```

let's activate it:

```python
disp.activate_for(res) # install for res (actually for type(res), but we infer that if what is given is an instance)
```

```python
res
```

We know have 1 click access to the content (if json), neatly formatted, and collapsible. we also have special cased the header field.

##  Beta feature

you can actually activate rich display for arbitrary container (making them collapsible), though this can have nasty side effects, so use at your own risk:

```python
from IPython.display import HTML
```

```python
ex = ({}, print, HTML, [{'hello':'world','This is a key':'and its value'}, 123456789], HTML('<b>BOLD</b>'), set())
ex
```

```python
disp.activate_for(list)
disp.activate_for(tuple)
disp.activate_for(set)
disp.activate_for(dict)
```
